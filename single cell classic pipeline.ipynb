```python
# ======================================================
# 1. IMPORT LIBRARIES
# ======================================================
import scanpy as sc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import harmonypy as hm
import os
import glob

sc.settings.verbosity = 3
sc.settings.set_figure_params(dpi=100)

# ======================================================
# 2. SET CORRECT DATA DIRECTORY
# ======================================================
data_dir = r"C:\Users\Shree\scRNA_project\data"

files = glob.glob(os.path.join(data_dir, "*.h5"))

print("Detected dataset files:")
for f in files:
    print(os.path.basename(f))

print("Total datasets:", len(files))

# ======================================================
# 3. LOAD AND MERGE ALL SAMPLES
# ======================================================
adatas = []

for file_path in files:
    file_name = os.path.basename(file_path)
    sample_name = file_name.split("_")[0]  # GSM ID
    
    ad = sc.read_10x_h5(file_path)
    ad.var_names_make_unique()
    
    ad.obs["sample"] = sample_name
    adatas.append(ad)

adata = sc.concat(adatas, label="batch", keys=[a.obs["sample"][0] for a in adatas])

print("Merged dataset shape:", adata.shape)

# ======================================================
# 4. QUALITY CONTROL
# ======================================================
adata.var["mt"] = adata.var_names.str.startswith("MT-")

sc.pp.calculate_qc_metrics(
    adata,
    qc_vars=["mt"],
    percent_top=None,
    log1p=False,
    inplace=True
)

sc.pp.filter_cells(adata, min_genes=200)
adata = adata[adata.obs.pct_counts_mt < 20, :]

print("After QC:", adata.shape)

# ======================================================
# 5. NORMALIZATION
# ======================================================
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)

# ======================================================
# 6. HIGHLY VARIABLE GENES (Batch-aware)
# ======================================================
sc.pp.highly_variable_genes(
    adata,
    flavor="seurat",
    n_top_genes=2000,
    batch_key="batch"
)

adata = adata[:, adata.var.highly_variable]

print("After HVG selection:", adata.shape)

# ======================================================
# 7. SCALING + PCA
# ======================================================
sc.pp.scale(adata, max_value=10)
sc.tl.pca(adata, svd_solver="arpack")

sc.pl.pca_variance_ratio(adata, log=True)

# ======================================================
# ======================================================
# 8. HARMONY BATCH CORRECTION (Corrected)
# ======================================================
ho = hm.run_harmony(adata.obsm['X_pca'], adata.obs, 'batch')

print("Harmony output shape:", ho.Z_corr.shape)
print("Original PCA shape:", adata.obsm['X_pca'].shape)

adata.obsm['X_pca'] = ho.Z_corr
# 9. NEIGHBORS + UMAP + CLUSTERING
# ======================================================
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=30)
sc.tl.umap(adata)
sc.tl.leiden(adata, resolution=0.5)

sc.pl.umap(adata, color=["leiden", "batch"])

# ======================================================
# 10. MARKER GENE ANALYSIS
# ======================================================
sc.tl.rank_genes_groups(adata, 'leiden', method='t-test')

sc.pl.rank_genes_groups(adata, n_genes=20, sharey=False)

markers = sc.get.rank_genes_groups_df(adata, group=None)
markers.to_csv("marker_genes.csv", index=False)

# ======================================================
# 11. SAVE FINAL DATA
# ======================================================
adata.write("integrated_scRNA_processed.h5ad")

print("Pipeline completed successfully.")
```
